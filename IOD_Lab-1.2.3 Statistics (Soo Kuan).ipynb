````python
# calculate statistical information of a set of observations using numpy and stats
observations = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100,10,32,49,10,10]
````

````python
import statistics
from scipy.stats import describe

def stats (x):
    x_np = np.array(x)
    if len(x_np) == 0:
        print("Error")
        return None
    mean_x = np.mean(x_np),
    median_x = np.median(x_np),
    mode_x = statistics.multimode(x_np)
    percentile_x = np.percentile(x_np, [25, 50,75])
    statistic_x = describe(x_np)

    summary = {
        "mean of observations" : mean_x,
        "median of observations" : median_x,
        "mode of observation" : mode_x,
        "percentile of observations" : percentile_x,
        "Statistical summary of observations" : statistic_x
    }

    return summary
        
````

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, se

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))
````
