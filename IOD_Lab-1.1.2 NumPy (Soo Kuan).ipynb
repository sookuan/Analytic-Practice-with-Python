## Synthetic Data

Sometimes we may want to generate test data, or we may need to initialise a series, matrix, or data frame for input to an algorithm. Numpy has several methods we can use for this.

Execute the following, then check the shape and content of each variable:

````python
# Creating arrays with initial values
a = np.zeros((3))
b = np.ones((1,3))
c = np.random.randint(1,10,(2,3,4))   # randint(low, high, size)
d = np.arange(4)
e = np.array( [[1,2,3,4], [5,6,7,8]] )
````

````python
### ANSWER
print(a, a.shape)
print(b, b.shape)
print(c, c.shape)
print(d, d.shape)
print(e, e.shape)
````

## Summary statistics with height, weight and population data

````python
np.random.seed(0)
height= np.round(np.random.normal(1.75, 0.20, 7000), 2)
weight= np.round(np.random.normal(60.32, 15, 7000), 2)
population=np.column_stack((height, weight))
````

````python
print('height:\n',height)
print('weight:\n',weight)
print('population:\n',population)
````

**1. Find Dimension of Array**

Find the dimension of each array: height, weight, population

````python
print("np.ndim(height):", np.ndim(height))       # 1D array
print("np.ndim(weight):", np.ndim(weight))       # 1D array
print("np.ndim(population):", np.ndim(population))  # 2D array
````

**10. Create histograms of height and weight**

````python
help(plt.hist)
````

````python
num_bins = 50
n, bins, patches = plt.hist(height, num_bins, facecolor ='blue', alpha=0.5)
plt.show()
````

````python
num_bins = 50
n, bins, patches = plt.hist(weight, num_bins, facecolor ='pink', alpha=0.5)
plt.show()
````

````python
# Create subplots: 1 row, 2 columns
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Histogram for Height
axs[0].hist(height, bins=5, color='blue', alpha=0.7)
axs[0].set_title('Height Distribution')
axs[0].set_xlabel('Height')
axs[0].set_ylabel('Frequency')
axs[0].grid(True)

# Histogram for Weight
axs[1].hist(weight, bins=5, color='green', alpha=0.7)
axs[1].set_title('Weight Distribution')
axs[1].set_xlabel('Weight')
axs[1].set_ylabel('Frequency')
axs[1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
````

**11. Create separate boxplots of height and weight**

````python
help(plt.boxplot)
````

````python
plt.boxplot(height)
plt.yscale('linear')
plt.grid(which ='major', alpha = 0.5, color='blue', linestyle = '--')
plt.grid(which ='minor', alpha =0.5, color='red', linestyle ='solid')
plt.show()
````

````python
plt.boxplot(weight)
plt.yscale('linear')
plt.grid(which ='major', alpha = 0.5, color='blue', linestyle = '--')
plt.grid(which ='minor', alpha =0.5, color='red', linestyle ='solid')
plt.show()
````

````python
# Create subplots for boxplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Boxplot for Height
axs[0].boxplot(height)
axs[0].set_title('Boxplot of Height')
axs[0].set_ylabel('Height')
axs[0].grid(True)

# Boxplot for Weight
axs[1].boxplot(weight)
axs[1].set_title('Boxplot of Weight')
axs[1].set_ylabel('Weight')
axs[1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
````
